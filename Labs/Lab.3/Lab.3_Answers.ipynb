* `cd ..`: changes current directory to the previous directory. Basically steps out of the current directory to the directory containing the current directory.
* `mkdir <directory>`: create a new directory with the specified name.
* `rmdir <directory>`: removes the specified directory. Note it has to be empty.
* `rm <filename>`: deletes the specified file.
* `mv <filename 1> <filename 2>`: Moves or renames a file.
* `cp <filename 1> <filename 2>`: copies an file. If you just provide a path to a directory, it copies the file into that directory with the same filename. If you specifiy a new filename, the copy has a new name. For example `cp File.1.txt File.2.txt` creates a copy of `File.1.txt` with the name `File.2.txt`. Meanwhile `cp File.1.txt my_directory`, where `my_directory` is a directory, creates a copy of `File.1.txt` in directory `my_directory` with the name `File.1.txt`.

````python
# Skeleton
def generate_uniform(N,x_min,x_max):
    out = []
    ### BEGIN SOLUTION
    
    for i in range(N):
        r = random.random()
        value = x_min + (x_max - x_min) * r
        out.append(value)

    ### END SOLUTION
    return out
````

Write a function that computes the mean of values in a list. Recall the equation for the mean of a random variable $\bf{x}$ computed on a data set of $n$ values $\{ x_i \} = \{x_1, x_2, ..., x_n\}$  is ${\bf\bar{x}} = \frac{1}{n} \sum_i^n x_i$.

````python
# Skeleton
def mean(Data):
    m=0.
    
    ### BEGIN SOLUTION

    total = 0.0
    n = len(Data)
    for value in Data:
        total = total + value
    if n > 0:
        m = total / n
    
    ### END SOLUTION
    
    return m
````

*Exercise 2b:* 
Write a function that computes the variance of values in a list. Recall the equation for the variance of a random variable $\bf{x}$ computed on a data set of $n$ values $\{ x_i \} = \{x_1, x_2, ..., x_n\}$  is ${\bf\langle x \rangle} = \frac{1}{n-1} \sum_i^n (x_i - {\bf\bar{x}})^2$.

````python
# Skeleton
def variance(Data):
    m=0.
    
    ### BEGIN SOLUTION

    n = len(Data)
    if n > 1:
        avg = mean(Data)
        total = 0.0
        for value in Data:
            diff = value - avg
            total = total + diff * diff
        m = total / (n - 1)     
    
    ### END SOLUTION
    
    return m
````

## Histogramming

*Exercise 3:* Write a function that bins the data so that you can create a histogram. An example of how to implement histogramming is the following logic:

* User inputs a list of values `x` and optionally `n_bins` which defaults to 10.
* If not supplied, find the minimum and maximum (`x_min`,`x_max`) of the values in x.
* Determine the bin size (`bin_size`) by dividing the range of the function by the number of bins.
* Create an empty list of zeros of size `n_bins`, call it `hist`.
* Loop over the values in `x`
    * Loop over the values in `hist` with index `i`:
        * If x is between `x_min+i*bin_size` and `x_min+(i+1)*bin_size`, increment `hist[i].` 
        * For efficiency, try to use continue to goto the next bin and data point.
* Return `hist` and the list corresponding of the bin edges (i.e. of `x_min+i*bin_size`).    

````python
# Solution
def histogram(x,n_bins=10,x_min=None,x_max=None):
    ### BEGIN SOLUTION

    # If min and max isn't given:
    if x_min is None:
        x_min = x[0]
        for value in x:
            if value < x_min:
                x_min = value
                
    if x_max is None:
        x_max = x[0]
        for value in x:
            if value > x_max:
                x_max = value
    
    # Determining bin size
    bin_size = (x_max - x_min) / n_bins
    
    # Create histogram list of zeros
    hist = []
    for i in range(n_bins):
        hist.append(0)
    
    # Create bin edges list
    bin_edges = []
    for i in range(n_bins):
        edge = x_min + i * bin_size
        bin_edges.append(edge)
    
    # Loop over data
    for value in x:
        # If value == x_max, put it in last bin
        if value == x_max:
            hist[n_bins - 1] = hist[n_bins - 1] + 1
            continue
        
        # Find correct bin
        for i in range(n_bins):
            lower = x_min + i * bin_size
            upper = x_min + (i + 1) * bin_size
            
            if value >= lower and value < upper:
                hist[i] = hist[i] + 1
                # Moves to next value
                break

    ### END SOLUTION

    return hist,bin_edges
````

```
[  6,  7] : #####
[  7,  8] : ######
[  8,  9] : ####

```

Where each line corresponds to a bin and the number of `#`'s are proportional to the value of the data in the bin. 

````python
# Solution
def draw_histogram(x,n_bins,x_min=None,x_max=None,character="#",max_character_per_line=20):
    ### BEGIN SOLUTION

    hist, bin_edges = histogram(x, n_bins, x_min, x_max)
    
    # Finds max count
    max_count = hist[0]
    for count in hist:
        if count > max_count:
            max_count = count
    
    # Loop through bins and print
    for i in range(n_bins):
        lower = bin_edges[i]
        
        # Compute upper edge
        if i < n_bins - 1:
            upper = bin_edges[i+1]
        else:
            upper = lower + (bin_edges[1] - bin_edges[0])
        
        count = hist[i]
        
        # Scales number of characters
        # [!!!] I didn't come up with this part of the solution myself. I didn't consider the max_character_per_line limit.
        # [!!!] An LLM helped me figure out the scaling problem, and I used its solution with tweaks
        if max_count > 0:
            scaled = int((count / max_count) * max_character_per_line)
        else:
            scaled = 0
        
        # Builds bar string
        bar = ""
        for j in range(scaled):
            bar = bar + character
        
        # Print formatted line
        # 6.2f limits to 6 characters, 2 decimal places as a floating point
        print(f"[{lower:6.2f}, {upper:6.2f}] : {bar}")      
    
    ### END SOLUTION

    return hist,bin_edges
````

If `p<=function(test_x)`, then place `test_x` on the output. If not, repeat the process, drawing two new numbers. Repeat until you have the specified number of generated numbers, `N`. For this problem, it's OK to determine the `min` and `max` by numerically sampling the function.  

````python
def generate_function(func,x_min,x_max,N=1000):
    out = list()
    ### BEGIN SOLUTION

    # I struggled a bit to understand the logic and asked an LLM to help explain how I'd approach this exercise
    
    # func is the function you want to sample from
    # x_min and x_max is the range where the sampling happens
    # N is how many numbers we want
    n_sample = 1000
    # Initialize max value to something value, which is the function value left of the boundary
    f_max = func(x_min)
    
    # Sample 1000 evenly spaced points
    for i in range(n_sample):
        # Creates evenly spaced x-values between x_min and x_max
        x = x_min + (x_max - x_min) * (i / n_sample)
        # Compute function at that point
        value = func(x)
        # If we found a bigger function value, we update the maximum
        if value > f_max:
            f_max = value

    # This keeps going until we've acacepted N points
    while len(out) < N:
        
        # Generates uniform random number between x_min and x_max
        test_x = x_min + (x_max - x_min) * random.random()
        # Generates random height number between 0 and f_max
        p = f_max * random.random()
        
        # If the random point is under the curve, we accept it
        if p <= func(test_x):
            # Store only accepted x-values
            out.append(test_x)
    
    ### END SOLUTION
    
    return out
````

*Exercise 9:* Combine your `generate_function`, `where`, and `in_range` functions above to create an integrate function. Use your integrate function to show that approximately 68% of Normal distribution is within one variance.

````python
def integrate(func, x_min, x_max, n_points=1000):
    
    # Sampling interval between +- 5 standard deviations
    sample_min = x_min -5
    sample_max = x_max +5
    
    # Generates distributed samples
    samples = generate_function(func, sample_min, sample_max, n_points)
    
    # Region test
    range_test = in_range(x_min, x_max)
    
    # Counts points in region
    indices = where(samples, range_test)
    
    # Monte Carlo probability
    integral = len(indices) / len(samples)
    
    return integral
````
