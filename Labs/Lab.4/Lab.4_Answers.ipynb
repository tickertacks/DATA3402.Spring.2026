1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        self.value = 0
        self.maximum = maximum
        
    def increment(self):
        if self.value < self.maximum:
            self.value = self.value + 1
        else:
            print("Error: Counter maxxed out!")
            
    def reset(self):
        self.value = 0
```

```python
# Length, width, x and y coordinates for corners
# Functions for area & perimeter
# All data members private w/accessors

class Rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    
    # Accessors
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    # Functions
    # Area = Legnth * Width
    def area(self):
        return self.__length * self.__width
    
    # Perimeter = (length * 2) + (width * 2)
    def perimeter(self):
        return self.__length * 2 + self.__width * 2
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
        
    # Accessors
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    # Not implemented functions (virtual)
    def area(self):
        raise NotImplementedError("Subclass must implement area method")
        
    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter method")
        
        
# ------------------------ New function ------------------------ <<<<<<<<<<<<
    def get_points(self):
        raise NotImplementedError("Subclass must implement get_points")
        
# ------------------------ Rectangle ------------------------
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        super().__init__(x, y)
    
    # Accessors
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    # Functions
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return self.__length * 2 + self.__width * 2

# ------------------------ New function ------------------------ <<<<<<<<<<<<
    
    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()
        
        # 4 points per side to get 16 pairs of x & y
        # Imagine a rectangular chocolate that's made of 9 smaller squares
        # Each intersecting line is a coordinate point
        for i in range(4):
            # Divides bottom and top edges of rectangle into 3 equal segments
            points.append((x + i * self.__length / 3, y))
            points.append((x + i * self.__length / 3, y + self.__width))
            # Divides left nad right edges of rectangle into 3 equal segments
            points.append((x, y + i * self.__width / 3))
            points.append((x + self.__length, y + i * self.__width /3))
            
        return points[:16]
    
# ------------------------ Circle ------------------------
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        super().__init__(x, y)
        
    # Accessors
    def get_radius(self):
        return self.__radius

    # Functions
    def area(self):
        return math.pi * self.__radius**2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius

# ------------------------ New function ------------------------ <<<<<<<<<<<<
    
    def get_points(self):
        points = []
        center_x = self.get_x()
        center_y = self.get_y()
        
        for i in range(16):
            # A circle is 2pi radians
            # Getting angle from 16 different slices
            angle = 2 * math.pi * i/16
            # Horizontal component
            x = center_x + self.__radius * math.cos(angle)
            # Vertical component
            y = center_y + self.__radius * math.sin(angle)
            points.append((x, y))
            
        return points

# ------------------------ Triangle ------------------------    

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        super().__init__(x, y)
    
    # Accessors
    def get_base(self):
        return self.__base
    
    def get_height(self):
        return self.__height
    
    def get_sides(self):
        return self.__side1, self.__side2, self.__side3
    
    # Functions
    def area(self):
        return 0.5 * self.__height * self.__base
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

# ------------------------ New function ------------------------ <<<<<<<<<<<<
    
    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()

        # 16 points -> 6/5/5 points per side
        # Triangle has points A, B, C
        A = (x, y)
        B = (x + self.__base, y)
        C = (x + self.__base / 2, y + self.__height)
        
        # x = x1 + t(x2 - x1)
        # y = y1 + t(y2 - y1)
        # t is the proportion from 0 to 1
        
        # If we have points A to B, then...
        # t = 0 -> Point A
        # t = 0.5 -> Mid-point
        # t = 1 -> Point B
        
        # A -> B
        for i in range(6):
            t = i/5
            px = A[0] + t * (B[0] - A[0])
            py = A[1] + t * (B[1] - A[1])
            points.append((px, py))
            
        # B -> C
        for i in range(5):
            t = i/4
            px = B[0] + t * (C[0] - B[0])
            py = B[1] + t * (C[1] - B[1])
            points.append((px, py))
            
        # C -> A
        for i in range(5):
            t = i/4
            px = C[0] + t * (A[0] - C[0])
            py = C[1] + t * (A[1] - C[1])
            points.append((px, py))
        
        return points[:16]
```

```python
class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
        
    # Accessors
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    # Not implemented functions (virtual)
    def area(self):
        raise NotImplementedError("Subclass must implement area method")
        
    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter method")
        
    def get_points(self):
        raise NotImplementedError("Subclass must implement get_points")
        
# ------------------------ New function ------------------------ <<<<<<<<<<<<
    def contains_point(self, px, py):
        raise NotImplementedError("Subclass must implement contains_point")
        
        
# ------------------------ Rectangle ------------------------
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        super().__init__(x, y)
    
    # Accessors
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    # Functions
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return self.__length * 2 + self.__width * 2

    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()
        
        for i in range(4):
            points.append((x + i * self.__length / 3, y))
            points.append((x + i * self.__length / 3, y + self.__width))
            points.append((x, y + i * self.__width / 3))
            points.append((x + self.__length, y + i * self.__width /3))
        return points[:16]

# ------------------------ New function ------------------------ <<<<<<<<<<<<
    
    def contains_point(self, px, py):
        x = self.get_x()
        y = self.get_y()
        
        # Checks for the point being both vertically and horizontally inside
        return (x <= px <= x + self.__length and
                y <= py <= y + self.__width)
        
# ------------------------ Circle ------------------------
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        super().__init__(x, y)
        
    # Accessors
    def get_radius(self):
        return self.__radius

    # Functions
    def area(self):
        return math.pi * self.__radius**2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def get_points(self):
        points = []
        center_x = self.get_x()
        center_y = self.get_y()
        
        for i in range(16):
            angle = 2 * math.pi * i/16
            x = center_x + self.__radius * math.cos(angle)
            y = center_y + self.__radius * math.sin(angle)
            points.append((x, y))
            
        return points
# ------------------------ New function ------------------------ <<<<<<<<<<<<

    def contains_point(self, px, py):
        # dx = horizontal distance from center
        # dy = vertical distance from center
        dx = px - self.get_x()
        dy = py - self.get_y()
        
        # r = sqrt(x^2+y^2)
        # Checks if the distance from the center less than or equal to the radius
        return math.hypot(dx, dy) <= self.__radius
    
# ------------------------ Triangle ------------------------    

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        super().__init__(x, y)
    
    # Accessors
    def get_base(self):
        return self.__base
    
    def get_height(self):
        return self.__height
    
    def get_sides(self):
        return self.__side1, self.__side2, self.__side3
    
    # Functions
    def area(self):
        return 0.5 * self.__height * self.__base
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()

        A = (x, y)
        B = (x + self.__base, y)
        C = (x + self.__base / 2, y + self.__height)

        # A -> B
        for i in range(6):
            t = i/5
            px = A[0] + t * (B[0] - A[0])
            py = A[1] + t * (B[1] - A[1])
            points.append((px, py))
            
        # B -> C
        for i in range(5):
            t = i/4
            px = B[0] + t * (C[0] - B[0])
            py = B[1] + t * (C[1] - B[1])
            points.append((px, py))
            
        # C -> A
        for i in range(5):
            t = i/4
            px = C[0] + t * (A[0] - C[0])
            py = C[1] + t * (A[1] - C[1])
            points.append((px, py))
        
        return points[:16]
    
# ------------------------ New function ------------------------ <<<<<<<<<<<<
# !!! I used an LLM to help me figure out the logic on how to check for points within a triangle using the method I used

    def contains_point(self, px, py):
        x = self.get_x()
        y = self.get_y()
        
        # Get vertices
        A = (x, y)
        B = (x + self.__base, y)
        C = (x + self.__base / 2, y + self.__height)
        
        # Imagine placing a point within a triangle to compare to
        # Our point P always divides the triangle into 3 different sections if you connect it to each vertex
        # If our point P /is/ within the triangle, it'd divide that triangle into sections that we can check
        # We'd check for equivalence of those 3 smaller triangles added to the 1 big initial triangle
        # If our point P /isn't/ within the triangle, then the area check would show it to be not equivalent to the initial triangle
        
        # Function to calculate area
        def area(x1, y1, x2, y2, x3, y3):
            return abs(x1 * (y2 - y3) +
                      x2 * (y3 - y1) +
                      x3 * (y1 - y2)) / 2
        
        # Total area of the full triangle
        total_area = area(A[0], A[1], B[0], B[1], C[0], C[1])
        
        # The 3 smaller triangles based on our Point
        area1 = area(px, py, B[0], B[1], C[0], C[1])
        area2 = area(A[0], A[1], px, py, C[0], C[1])
        area3 = area(A[0], A[1], B[0], B[1], px, py)
        
        # 0.001 allows tiny errors
        return abs(total_area - (area1 + area2 + area3)) < 0.001
```

```python
class RasterDrawing:
    def __init__(self, width, height):
        # Canvas creation
        self.canvas = Canvas(width, height)
        # List to hold shapes
        self.shapes = []
        
    # Adding a shape to drawing
    def add_shape(self, shape):
        self.shapes.append(shape)
        
    # Remove a shape from the drawing
    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)
            
    # Paints all shapes on canvas
    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
        self.canvas.display()
        
    # Clear drawing
    def clear(self):
        self.shapes = []
        self.canvas.clear_canvas()
```

```python
import math

# ------------------------ Canvas class from lecture ------------------------ #
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        
# ------------------------------------------------------------------------ #

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
        
    # Accessors
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    # Not implemented functions (virtual)
    def area(self):
        raise NotImplementedError("Subclass must implement area method")
        
    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter method")
        
    def get_points(self):
        raise NotImplementedError("Subclass must implement get_points")

    def contains_point(self, px, py):
        raise NotImplementedError("Subclass must implement contains_point")

    def overlaps(self, other):
        for px, py in self.get_points():
            if other.contains_point(px, py):
                return True
            
        for px, py in other.get_points():
            if self.contains_point(px, py):
                return True

        return False
    
    def paint(self, canvas, char="*"):
        for px, py in self.get_points():
            canvas.set_pixel(int(round(py)), int(round(px)), char)
    
# ------------------------ Rectangle ------------------------
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        super().__init__(x, y)
    
    # Accessors
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    # Functions
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return self.__length * 2 + self.__width * 2

    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()
        
        for i in range(4):
            points.append((x + i * self.__length / 3, y))
            points.append((x + i * self.__length / 3, y + self.__width))
            points.append((x, y + i * self.__width / 3))
            points.append((x + self.__length, y + i * self.__width /3))
        return points[:16]
    
    def contains_point(self, px, py):
        x = self.get_x()
        y = self.get_y()
        
        return (x <= px <= x + self.__length and
                y <= py <= y + self.__width)
        
# ------------------------ New Code ------------------------ <<<<<<<<<<<<

    def __repr__(self):
        return (f"Rectangle({self.get_length()}, {self.get_width()}, {self.get_x()}, {self.get_y()})")
        
# ------------------------ Circle ------------------------
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        super().__init__(x, y)
        
    # Accessors
    def get_radius(self):
        return self.__radius

    # Functions
    def area(self):
        return math.pi * self.__radius**2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def get_points(self):
        points = []
        center_x = self.get_x()
        center_y = self.get_y()
        
        for i in range(16):
            angle = 2 * math.pi * i/16
            x = center_x + self.__radius * math.cos(angle)
            y = center_y + self.__radius * math.sin(angle)
            points.append((x, y))
            
        return points

    def contains_point(self, px, py):
        dx = px - self.get_x()
        dy = py - self.get_y()
        
        return math.hypot(dx, dy) <= self.__radius
    
# ------------------------ New Code ------------------------ <<<<<<<<<<<<

    def __repr__(self):
        return (f"Circle({self.get_radius()}, {self.get_x()}, {self.get_y()})")
    
# ------------------------ Triangle ------------------------    

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        super().__init__(x, y)
    
    # Accessors
    def get_base(self):
        return self.__base
    
    def get_height(self):
        return self.__height
    
    def get_sides(self):
        return self.__side1, self.__side2, self.__side3
    
    # Functions
    def area(self):
        return 0.5 * self.__height * self.__base
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_points(self):
        points = []
        x = self.get_x()
        y = self.get_y()

        A = (x, y)
        B = (x + self.__base, y)
        C = (x + self.__base / 2, y + self.__height)

        # A -> B
        for i in range(6):
            t = i/5
            px = A[0] + t * (B[0] - A[0])
            py = A[1] + t * (B[1] - A[1])
            points.append((px, py))
            
        # B -> C
        for i in range(5):
            t = i/4
            px = B[0] + t * (C[0] - B[0])
            py = B[1] + t * (C[1] - B[1])
            points.append((px, py))
            
        # C -> A
        for i in range(5):
            t = i/4
            px = C[0] + t * (A[0] - C[0])
            py = C[1] + t * (A[1] - C[1])
            points.append((px, py))
        
        return points[:16]

    def contains_point(self, px, py):
        x = self.get_x()
        y = self.get_y()
        
        # Get vertices
        A = (x, y)
        B = (x + self.__base, y)
        C = (x + self.__base / 2, y + self.__height)
        
        def area(x1, y1, x2, y2, x3, y3):
            return abs(x1 * (y2 - y3) +
                      x2 * (y3 - y1) +
                      x3 * (y1 - y2)) / 2
        
        total_area = area(A[0], A[1], B[0], B[1], C[0], C[1])
        
        area1 = area(px, py, B[0], B[1], C[0], C[1])
        area2 = area(A[0], A[1], px, py, C[0], C[1])
        area3 = area(A[0], A[1], B[0], B[1], px, py)
        
        return abs(total_area - (area1 + area2 + area3)) < 0.001

# ------------------------ New Code ------------------------ <<<<<<<<<<<<

    def __repr__(self):
        return (f"Triangle({self.__base}, {self.__height}, {self.__side1}, {self.__side2}, {self.__side3}, {self.get_x()}, {self.get_y()})")
    
# ------------------------ CompoundShape ------------------------

class CompoundShape(Shape):
    def __init__(self, shapes=[]):
        super().__init__(0, 0)
        self.shapes = shapes
        
    def add_shape(self, shape):
        self.shapes.append(shape)
        
    def paint(self, canvas, char="*"):
        for shape in self.shapes:
            shape.paint(canvas, char)

# ------------------------ New Code ------------------------ <<<<<<<<<<<<
            
    def __repr__(self):
        shapes_repr = ",".join([repr(s) for s in self.shapes])
        return (f"CompoundShape([{shapes_repr}])")

# ------------------------ RasterDrawing ------------------------
    
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []
        
    def add_shape(self, shape):
        self.shapes.append(shape)
        
    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
        self.canvas.display()
        
    def clear(self):
        self.shapes = []
        self.canvas.clear_canvas()

# ------------------------ New Code ------------------------ <<<<<<<<<<<<

    def save(self, filename):
        # Opens file for writing with the file object
        with open(filename, "w") as f:
            # Writes first line to store canvas dimensions
            f.write(f"{self.canvas.width}, {self.canvas.height}\n")
            # Loops through all shapes in drawings
            for shape in self.shapes:
                # Turns into constructor string and writes to file
                f.write(repr(shape) + "\n")
    
# ------------------------ Loader ------------------------
    
def raster_loader(filename):
    # Opens file for reading
    with open(filename, "r") as f:
        # Reads dimension
        first_line = f.readline().strip()
        # Converts strings into integers
        width, height = map(int, first_line.split(","))
        
        # Recreates blank drawing with the canvas size
        drawing = RasterDrawing(width, height)

        # Reads each remaining line
        for line in f:
            # Removes whitespace and newlines
            line = line.strip()
            # Ignores empty lines
            if line:
                # If the line matches the shape, it gets evaluated and then called as that object
                shape = eval(line)
                drawing.add_shape(shape)

    # Drawing finally rebuilt
    return drawing

# NOTES:
# __repr__ gives a string representation of an object to return
# With CompoundShape, it makes a list of the string representation of shapes
# We essentially reconstruct stuff
```
